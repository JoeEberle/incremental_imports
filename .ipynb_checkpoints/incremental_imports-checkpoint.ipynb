## Step 1  - Read in Pandas only recent data 

```python
import pandas as pd

def process_and_append_records(source_file, start_date, df_emr_all_time):
    # Read the source CSV file into df_emr_current
    df_emr_current = pd.read_csv(source_file)

    # Convert 'record_date' column to datetime format
    df_emr_current['record_date'] = pd.to_datetime(df_emr_current['record_date'])

    # Filter records based on the specified start_date
    df_emr_current = df_emr_current[df_emr_current['record_date'] > start_date]

    # Append records to df_emr_all_time and remove duplicates
    df_emr_all_time = pd.concat([df_emr_all_time, df_emr_current]).drop_duplicates().reset_index(drop=True)

    return df_emr_all_time

# Example usage:
source_file_path = '/path/to/source_file.csv'
start_date = pd.to_datetime('2023-01-01')  # Replace with your specific start date
df_emr_all_time = pd.DataFrame()  # Initialize an empty DataFrame

# Call the function to process and append records
df_emr_all_time = process_and_append_records(source_file_path, start_date, df_emr_all_time)

# Print the resulting DataFrame
print(df_emr_all_time)
```
